```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
df = pd.read_excel('/content/2.simple_graph.xlsx')
```

```python
df.head()
```

```python
fig = plt.figure(figsize=(6,4))
axes = fig.add_subplot(111)

sns.lineplot(x = 'Item_1', y = 'Item_1', data=df, ax=axes)
sns.lineplot(x = 'Item_1', y = 'Item_2', data=df, ax=axes)
sns.lineplot(x = 'Item_1', y = 'Price', data=df, ax=axes)
sns.lineplot(x = 'Item_1', y = 'Age', data=df, ax=axes)
```

```python
fig = plt.figure(figsize=(6,10))

axes = fig.add_subplot(411)

sns.lineplot(x = 'Item_1', y = 'Item_1', data=df, ax=axes)

axes = fig.add_subplot(412)
sns.lineplot(x = 'Item_1', y = 'Item_2', data=df, ax=axes)

axes = fig.add_subplot(413)
sns.lineplot(x = 'Item_1', y = 'Price', data=df, ax=axes)

axes = fig.add_subplot(414)
sns.lineplot(x = 'Item_1', y = 'Age', data=df, ax=axes)

plt.show()
```

```python
df1 = df[['Item_3','Price']].groupby('Item_3', as_index=False).agg('sum').sort_values('Price', ascending=False).reset_index(drop=True)
df1
```

```python
sns.barplot(x ='Item_3', y ='Price', data = df1, palette ='plasma')
```
