```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8,8))

df1.plot(kind='bar', ax=axes[0,0])
axes[0,0].set_xlabel('x-axis')
axes[0,0].set_ylabel('y-axis')
axes[0,0].set_title("Graphs showing y = x and y = x**2", fontsize=15)


df1.plot(kind='barh', ax=axes[0,1])
axes[0,1].set_xlabel('x-axis')
axes[0,1].set_ylabel('y-axis')
axes[0,1].set_title("Graphs showing y = x and y = x**2", fontsize=15)


df['Age'].plot(kind='hist', bins=20, ax=axes[1,0])
axes[1,0].set_title('Age Distribution', fontsize=15)
axes[1,0].set_xlabel('Age')
axes[1,0].set_ylabel('Value')


df['Age'].plot(kind='box', ax=axes[1,1])
axes[1,1].set_title('Age Distribution', fontsize=15)
axes[1,1].set_xlabel('Age')
axes[1,1].set_ylabel('Value')


plt.tight_layout()
```
