```python
feature_cols = ['Gender', 'Coding', 'Math', 'SoloWork', 'Apply Program', 'Marketing', 'Acting', 'Drawing',
                'Curiour', 'Present', 'Negotiate', 'Problems Solving', 'Offer ideas', 'Think outside the box', 'Calculate']
```

```python
X = df[feature_cols]
y = df['Department']
```

```python
X.dtypes
```

```python
X['Gender'].value_counts()
```

```python
def convert_gender(x):
    if x == 'M':
        return 1
    else:
        return 0
```

```python
X['Gender'] = X['Gender'].map(convert_gender)
```

```python
X.head()
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier(max_depth=4)
clf = clf.fit(X,y)
```
