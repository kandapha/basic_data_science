### __<font color='blue'>Data Exploratory (Pivot Table)</font>__

<font color='red'>1. แต่ละ Brand ขายสินค้าได้เท่าใด // <br>หาผลรวมของสินค้าแบ่งตามรายชื่อ Product (Brand) // หรือแต่ละ Brand ขายสินค้าได้เท่าไหร่</font>

```python
df.groupby('Product')[['Price']].agg('sum')
```

<font color='red'>2. แต่ละ Brand ขายสินค้าได้เท่าใด <br>เรียงลำดับผลรวมราคาสินค้า แบ่งตาม Brand</font>

```python
df.groupby('Product')[['Price']].agg('sum').sort_values(by='Price', ascending=False)
```

<font color='red'>3. แปลว่า Nike คือ ฺBrand ที่คนนิยมซื้อมากที่สุดจริงหรือไม่ <br>
ให้พิจารณาจากผลรวมของ Price และ Unit ประกอบด้วย
</font>

สินค้าที่คนนิยมซื้อมากที่สุดคือ __Nike__เมื่อพิจารณาจากราคารวมที่สูงที่สุด)

แต่ถ้าพิจารณาจากจำนวน คนจะนิยมซื้อ Brand __H&M__ มากกว่า

```python
df.groupby('Product')[['Price','Unit']].agg('sum').sort_values(by='Price', ascending=False)
```

<font color='red'>4. สินค้าประเภทใดที่คนนิยมซื้อมากที่สุด</font> <br> เมื่อพิจารณาจาก ประเภทสินค้า ร้านค้ามียอดขายมากที่สุด __Shoes__ 
<br> แต่คนมักนิยมซื้อ __เสื้อ__ มีจำนวนมากที่สุด


```python
df.groupby('Category')[['Price','Unit']].agg('sum').sort_values(by='Price', ascending=False)
```

<font color='red'>5. แสดงยอดขายและจำนวน แบ่งตาม Brand และ Category</font>


```python
p_table = pd.pivot_table(df, index = ['Product','Category']) # default aggfunc = mean
p_table
```

```python
p_table = pd.pivot_table(df, index = ['Product','Category'],
                         aggfunc= {'Price': 'sum', 'Unit': 'sum'})
p_table
```

<font color='blue'>สรุป</font> <br>
- Brand ที่ขายได้ดีที่สุดในร้าน คือ __Nike__

สินค้าที่ขายได้ยอดเงินเยอะที่สุดในแต่ละ Brand

ดังนั้นร้านควร
1. วางขายสินค้า ประเภทกางเกงขาสั้น รองเท้ากีฬาเยอะๆ

```python
p_table = pd.pivot_table(df, index = ['Product'], values = ['Price','Unit'], 
                         aggfunc= {'Price': ['sum','mean','max','min'] ,
                                   'Unit' : ['sum','count']})
p_table
```

```python
p_table.columns = p_table.columns.droplevel(0)
p_table
```

```python
#p_table.columns.name = None
#p_table
```

```python
p_table = p_table.reset_index()
p_table
```

```python
p_table.iloc[: , -1]  #U_sum
```

```python
p_table.iloc[: , 4] #P_sum
```

```python
p_table['U_sum'] = p_table.iloc[: , -1].to_list()
```

```python
p_table['P_sum'] = p_table.iloc[: , 4].to_list()
```

```python
p_table
```

```python
p_table = p_table.drop('sum', axis=1)
#p_table = p_table.drop(p_table.columns[4], axis=1)
```

```python
p_table
```

```python
p_table = p_table.rename(columns={ p_table.columns[1]: 'P_max',
                         p_table.columns[2]: 'P_mean',
                         p_table.columns[3]: 'P_min',
                         p_table.columns[4]: 'U_count'})
p_table
```

```python
p_table['price per unit'] = p_table['P_sum'] / p_table['U_sum']
p_table
```
