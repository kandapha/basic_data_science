###แบบฝึกหัด แสดงสูตรคูณ

```python
def line():
  print('-'*30)

def mult_table(N,M):
  for n in range(N,M+1):
    for i in range(1,13):
      print('{} x {} = {}'.format(n,i,n*i))
      i+=1
    n+=1
    line()

N= int(input('Enter number1 : '))
M= int(input('Enter number2 : '))
line()
mult_table(N,M)
```

แบบฝึกหัด แปลงอุณหภูมิ 

```python
def temp(C):
    F= 9 * float(C) /5 +32
    K = float(C) + 273.15
    return  F,K

for x in range(1,21):
    A=temp(x)
    #A= A[0],A[1]
    print("{:2d} Celsius = {} farenheit = {} kelvin".format(x,A[0],A[1]))

```

##WHILE LOOP

```python
#While Loop
def while_demo():
    i = 1
    total = 0
    
    while True:
        num = int(input("Enter number {0} : ".format(i)))
        if num != 0:
            total+=num
            i+=1
        else:
            break
    print("Total = {0}".format(total))
while_demo()
```

###แบบฝึกหัด คำนวณค่ามากที่สุดและค่าน้อยที่สุด

###แบบฝึกหัด คำนวณค่าเสื่อมราคารถยนต์

```python
def line():
  print( '-'*50)
Price = float(input("กรอกราคารถที่ซื้อมา : "))
Depreciation = float(input("กรอกค่าเสื่อมราคาเป็น % ต่อปี : "))
line()
i = 0
half = Price/2

while True:
    i+=1
    Falling = Price * (Depreciation/100)
    Price = Price - Falling
    print("ปีที่ {:2} ราคาลดลง {:12,.2f} บาท เหลือ {:12,.2f} บาท".format(i,Falling,Price))
    if Price < half:
        break
line()
```

```python
numbers = [1, 2, 3, 4, 5]
numbers[0] # returns the first item 
numbers[1] # returns the second item
numbers[-1] # returns the first item from the end
numbers[-2] # returns the second item from the end 
numbers.append(6) # adds 6 to the end
numbers.insert(0, 6) # adds 6 at index position of 0
numbers.remove(6) # removes 6
numbers.pop() # removes the last item
numbers.clear() # removes all the items
##numbers.index(8) # returns the index of first occurrence of 8
numbers.sort() # sorts the list
numbers.reverse() # reverses the list
numbers.copy() # returns a copy of the list 
```
