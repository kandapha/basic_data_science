# __Subplot__

```python
fig, axes = plt.subplots(nrows=3,ncols=2,figsize=(10,10))

x = df['Item_1'].tolist()
y = df['Item_2'].tolist()
axes[0,0].plot(x, x, 'r--', linewidth=5)
axes[0,0].plot(x, y, 'g-o')
axes[0,0].set_xlabel('x-axis')
axes[0,0].set_ylabel('y-axis')
axes[0,0].set_title("Graphs showing y = x and y = x**2", fontsize=15)

x = df1.index.values.tolist()
y = df1['Price'].values.tolist()
axes[0,1].barh(x,y)
axes[0,1].set_title('Sales by category', fontsize=15)
axes[0,1].set_ylabel('Sales revenue')
axes[0,1].set_xlabel('Item type')

axes[1,0].scatter(df['Age'] , df['Price'])
axes[1,0].set_title('Age Distribution', fontsize=15)
axes[1,0].set_xlabel('Age')
axes[1,0].set_ylabel('Value')

axes[1,1].boxplot(df['Age'])
axes[1,1].set_title('Age Distribution', fontsize=15)
axes[1,1].set_xlabel('Age')
axes[1,1].set_ylabel('Value')

df2 = df[df.Age<70]
x = df2['Age']

axes[2,0].scatter(df2['Age'] , df2['Price'])
axes[2,0].set_title('Age Distribution', fontsize=15)
axes[2,0].set_xlabel('Age')
axes[2,0].set_ylabel('Value')

axes[2,1].boxplot(x)
axes[2,1].set_title('Age Distribution', fontsize=15)
axes[2,1].set_xlabel('Age')
axes[2,1].set_ylabel('Value')

plt.tight_layout()
```
